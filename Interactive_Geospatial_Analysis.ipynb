```python
!pip install folium geopandas

import folium
import geopandas as gpd
import pandas as pd
from folium.plugins import HeatMap

# Create interactive map
def create_crime_heatmap():
    # Sample crime data (latitude, longitude, weight)
    crime_data = [
        [40.7128, -74.0060, 0.8],  # NYC
        [34.0522, -118.2437, 0.6], # LA
        [41.8781, -87.6298, 0.7],  # Chicago
        [29.7604, -95.3698, 0.5],  # Houston
        [33.4484, -112.0740, 0.4]  # Phoenix
    ]

    # Create base map
    m = folium.Map(location=[39.8283, -98.5795], zoom_start=4)

    # Add heatmap
    HeatMap(crime_data).add_to(m)

    # Add markers
    for lat, lon, weight in crime_data:
        folium.CircleMarker(
            location=[lat, lon],
            radius=weight * 20,
            popup=f'Crime intensity: {weight}',
            color='red',
            fill=True
        ).add_to(m)

    return m

# Display map
crime_map = create_crime_heatmap()
crime_map
```
